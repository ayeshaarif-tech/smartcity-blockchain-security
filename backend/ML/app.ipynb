```python
from flask import Flask, request, render_template
import pickle
import pandas as pd

app = Flask(__name__)

@app.route("/")
def home():
    return render_template("index.html")

@app.route("/predict", methods=['POST'])
def predict():
    # Load the trained model
    with open('random_forest_model.pkl', 'rb') as f:
        model = pickle.load(f)
    
    # Get the form data
    solar_rad = float(request.form['solar_rad'])
    dewpt = float(request.form['dewpt'])
    uv = float(request.form['uv'])
    wind_dir = int(request.form['wind_dir'])
    precip = float(request.form['precip'])
    pop = int(request.form['pop'])
    ozone = int(request.form['ozone'])
    app_temp = float(request.form['app_temp'])

    # Prepare input data for prediction
    input_data = pd.DataFrame([[solar_rad, dewpt, uv, wind_dir, precip, pop, ozone, app_temp]],
                              columns=['solar_rad', 'dewpt', 'uv', 'wind_dir', 'precip', 'pop', 
                                       'ozone', 'app_temp'])

    # Make prediction
    prediction = model.predict(input_data)[0]

    # Display prediction result
    prediction_text = f"The predicted weather description is: {prediction}"

    return render_template('index.html', prediction_text=prediction_text)

if __name__ == "__main__":
    app.run(debug=True)

```
